```python
from catmap import ReactionModel
```

```python
import numpy as np
```

```python
mkm_file = 'example_network.mkm'
model = ReactionModel(setup_file=mkm_file)
```

```python
scaler_variables = ['free_energy','zero_point_energy','enthalpy','entropy']
solver_variables = ['production_rate', 'coverage', 'rate', # 'rate_control',
                    'consumption_rate', 'rate_constant', 'equilibrium_constant',
                    'rxn_order', 'apparent_activation_energy',
                    'directional_rates', 'forward_rate', 'reverse_rate',
                    'forward_rate_constant', 'reverse_rate_constant']
two_dim_variables = ['rate_control', 'rxn_order']
one_dim_variables = list(set(scaler_variables + solver_variables) - set(two_dim_variables))
model.output_variables += solver_variables + scaler_variables
```

```python
model.run()
```

```python
from catmap import analyze
```

```python
plot_variables = ['production_rate'] #['free_energy','enthalpy','entropy']
```

```python
plot_variables =   ['production_rate', 'coverage', 'rate',
                    'consumption_rate', 'rate_constant', 
                    'directional_rates', 'forward_rate', 'reverse_rate',
                    'forward_rate_constant', 'reverse_rate_constant']
```

```python
for var in plot_variables:
    if var in one_dim_variables:
        vm = analyze.VectorMap(model)
    if var in two_dim_variables:
        vm = analyze.MatrixMap(model)
    vm.plot_variable = var  # tell the model which output to plot
    vm.log_scale = True  # rates should be plotted on a log-scale
    xy,z = vm.get_pts_cols()
    z = np.array(z)
    vm.min = 1e-22  # minimum rate to plot
    vm.max = z.max()  # maximum rate to plot
    vm.threshold = 1e-25  # anything below this is considered to be 0
    vm.unique_only = False
    vm.subplots_adjust_kwargs = {'left': 0.2, 'right': 0.8, 'bottom': 0.15}
    fig = vm.plot(save=var+'.pdf')
    fig.clf()
```

```python
ma = analyze.MechanismAnalysis(model)
ma.energy_type = 'free_energy' #can also be free_energy/potential_energy
ma.include_labels = False #way too messy with labels
ma.pressure_correction = False #assume all pressures are 1 bar (so that energies are the same as from DFT)
ma.include_labels = True
```

```python
fig = ma.plot(save='FED.png')
```
